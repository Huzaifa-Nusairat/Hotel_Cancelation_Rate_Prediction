```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_train = torch.from_numpy(X_train.astype(np.float32))
X_val = torch.from_numpy(X_val.astype(np.float32))
```

```python
train_dataset = TensorDataset(X_train,y_train) 
val_dataset =  TensorDataset(X_val,y_val)
```

```python
batch_size = 32
# n_iters = 10000
# numberOfEpochs = n_iters / (len(train_dataset) / batch_size)
# numberOfEpochs = int(numberOfEpochs)
numberOfEpochs = 300
train_loader = DataLoader(train_dataset,batch_size = batch_size , shuffle = False, drop_last=True)
val_loader = DataLoader(val_dataset, batch_size = val_dataset.tensors[0].shape[0]  ,shuffle = False)
print(numberOfEpochs)
```

```python
class ANNModel(nn.Module):
    
    def __init__(self, input_dim, hidden_dim1, hidden_dim2, output_dim):
        super(ANNModel, self).__init__()
        
        self.fc1 = nn.Linear(input_dim, hidden_dim1) 
        self.selu1 = nn.ReLU()
        self.dropOut1 = nn.Dropout(p=0.4)
        
        self.fc2 = nn.Linear(hidden_dim1, hidden_dim2)
        self.selu2 = nn.ReLU()       
        self.dropOut2 = nn.Dropout(p=0.3)
        

        self.fc4 = nn.Linear(hidden_dim2, output_dim)
    
    def forward(self, x):
        
        out = self.fc1(x)
        out = self.selu1(out)
        out = self.dropOut1(out)
        
        out = self.fc2(out)
        out = self.selu2(out)
        out = self.dropOut2(out)
        
        out = self.fc3(out)
        out = self.selu3(out)
        out = self.dropOut3(out)
        
        
        out = self.fc4(out)
        return out
input_dim = 32
hidden_dim1 = 512
hidden_dim2 = 256
output_dim = 1
model = ANNModel(input_dim, hidden_dim1, hidden_dim2, output_dim)
error = nn.BCEWithLogitsLoss()
learning_rate = 0.0001
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
train_loss = torch.zeros(numberOfEpochs)
val_loss  = torch.zeros(numberOfEpochs)
train_accurecy  = torch.zeros(numberOfEpochs)
val_accurecy   = torch.zeros(numberOfEpochs)

val_df = pd.DataFrame()
true_values = []
pre_values = []
for epoch in range(numberOfEpochs):
    batchLoss = []
    for X_train, Y_train in train_loader:
        output = model(X_train)
        loss = error(output, Y_train)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        batchLoss.append(loss.item())
        train_predict = (torch.sigmoid(output)>.5).float()
        train_accurecy[epoch] = 100*torch.mean((train_predict == Y_train).float())
    
    train_loss[epoch] = np.mean(batchLoss)
    X_val, Y_val = next(iter(val_loader))
    val_output = model(X_val)
    loss = error(val_output,Y_val)
    val_loss[epoch] = loss.item()
    val_predict = (torch.sigmoid(val_output)>.5).float()
    val_accurecy[epoch] = 100*torch.mean((val_predict == Y_val).float())
    print("Epoch: {}\tTrain_Loss: {}\tTrain_Accurecy: {}\tValidation_Loss: {}\tValidation_Accurecy: {}\t".format(epoch, train_loss[epoch], train_accurecy[epoch], val_loss[epoch], val_accurecy[epoch]))
for i in range(len(Y_val)):
    true_values.append(Y_val[i].numpy().tolist()[0])
    pre_values.append(val_predict[i].numpy().tolist()[0])
val_df["True_Values"] = true_values
val_df["Predicted_Values"] = pre_values
```

```python
fig,ax = plt.subplots(1,2,figsize=(16,5))

ax[0].plot(train_loss,'s-',label='Train')
ax[0].plot(val_loss,'s-',label='Validation')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].set_title('Model loss')
ax[0].legend()

ax[1].plot(train_accurecy,'s-',label='Train')
ax[1].plot(val_accurecy,'o-',label='Validation')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy (%)')
ax[1].set_title(f'Final model Validation accuracy: {val_accurecy[-1]:.2f}%')
ax[1].legend()
plt.show()
```

```python
def Classification_Report(valid_target, pred):
    classes = ["0", "1"]
    length = len(classes)
    print('F1-Score: ', f1_score(valid_target, pred, average='weighted'))
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))
    clf_report = classification_report(valid_target,
                                   pred,
                                   output_dict=True)
    g = sns.heatmap(pd.DataFrame(clf_report).iloc[:-1, :].T, annot=True, cmap='Blues', ax=ax1)
    g.set_title('Classification Report\n\n');
    #=========================================
    cf_matrix = confusion_matrix(valid_target, pred)
    g1 = sns.heatmap(cf_matrix, annot=True, cmap='Blues',ax=ax2)
    g1.set_title('Confusion Matrix\n\n');
    g1.set_xlabel('\nPredicted Values')
    g1.set_ylabel('Actual Values ');
    plt.xticks(np.arange(length)+.5, classes, rotation= 90)
    plt.yticks(np.arange(length)+.5, classes, rotation=0)
    
    plt.show()
```
